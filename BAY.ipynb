```python
#Средняя цена покупки
rs = requests.get(Url)

m = re.search(r'var line1=(.+);', rs.text)

data_str = m.group(1)

data = json.loads(data_str)
#for i in data:
#    print(i)


def first_way(data):
    df = pd.DataFrame(data, columns=['date', 'price', 'volume'])
    
    # Preprocess the data
    df['date'] = df['date'].str.replace(': +0', '')
    df['date'] = pd.to_datetime(df['date'], format='%b %d %Y %H')
    df = df.set_index('date')
    
    # Calculate minimum and maximum prices for the last 6 months
    n_periods = 50  # Number of days in 6 months
    min_price = df['price'][-n_periods:].min()
    max_price = df['price'][-n_periods:].max()
    
    
    mp = min_price + (min_price * 0.10)
    mx = max_price - (max_price * 0.10)
    #print(f'Minimum price : {min_price} USD')
    #print(f'Maximum price : {max_price} USD')
    #print(f'-10% lower maximum price: {mx} USD')
    #print(f'+10% lower minimum price: {mp} USD')
    
    mm = (mx - (mx * 0.13)) - mp
    #print(f'mm value: {mm} USD')
    
    new_value = mp + mm
    
    percentage_difference = ((new_value - mp) / mp) * 100
    
    #print(f'Percentage difference: {percentage_difference:.2f}%')
    
    pricess = [float(x[1]) for x in data]

    # вычисляем среднее значение цены
    avg_price = statistics.mean(pricess[-120:])
    avg_price_10 = avg_price - (avg_price * 0.10)
    #print(f'Средняя цена:  {avg_price_10:.2f} USD ')
    currency_rates = get_currency_rates('USD',avg_price_10)
    price = currency_rates['RUB']
    #print(f'Перевели: {avg_price_10} USD в',price,'RUB')

    # извлекаем количество проданных единиц товара из данных
    #quantities = [float(x[2]) for x in data]

    # вычисляем среднее количество проданных единиц товара
    #avg_quantity = statistics.mean(quantities)

    #print(f'Среднее количество проданных единиц товара: {avg_quantity:.2f} USD ')
    return price
    
price = first_way(data)
print(f'Средний прайс покупки равен:{price} RUB')
```

```python
def chek_Personality(driver, link):
    print("chek_Personality")
    driver.get(link)
    try:
        div_elements_message = driver.find_element(By.CSS_SELECTOR, 'div.market_listing_table_message')
        text_message = "Ошибка поиска. Повторите попытку позднее."
        div_elements_message_text = div_elements_message.text
        if div_elements_message_text == text_message:
            driver.refresh()
            opis_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, '//a[text()="Описание"]'))
            )
    except NoSuchElementException:
        opis_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//a[text()="Описание"]'))
        )

    try:
        opis_element.click()

        # Wait for a short period to ensure the content is loaded
        wait_time = random.uniform(1, 2)
        time.sleep(wait_time)

        # Find the element by its ID
        element = driver.find_element(By.ID, 'sih_block_best_offer_of_marketplace_tabs_1')

        # Get the HTML content of the element
        html_content = element.get_attribute('outerHTML')

        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(html_content, 'html.parser')

        # Find all div elements with the class "descriptor"
        descriptors = soup.find_all('div', class_='descriptor')

        # Iterate through the descriptors to find the specific text
        for descriptor in descriptors:
            if "Включает личность" in descriptor.get_text():
                driver.back()
                return descriptor.get_text()

        # Return None if the text is not found
        driver.back()
        return None

    except NoSuchElementException:
        driver.back()
        return None
```
